```python
print(df["Year"].unique())
print(df["Year"].dtype)
```

```python
tech = df[df["Category"] == "Technology"]
ventas_anuales = tech.groupby("Year")["Sales"].sum()
print(ventas_anuales)
```

```python
import matplotlib.pyplot as plt
```

```python
ventas_anuales.plot(kind="line", marker="o", figsize=(10,5), color="darkorange")
plt.title("Evolución anual de ventas — Technology")
plt.xlabel("Año")
plt.ylabel("Ventas ($)")
plt.grid(True)
plt.show()
```

```python
import numpy as np

x = np.arange(len(ventas_anuales))
y = ventas_anuales.values

coeficientes = np.polyfit(x, y, 1)
tendencia = np.poly1d(coeficientes)

plt.figure(figsize=(10,5))
plt.plot(ventas_anuales.index, y, marker="o", label="Ventas reales", color="darkorange")
plt.plot(ventas_anuales.index, tendencia(x), linestyle="--", label="Línea de tendencia", color="red")
plt.title("Ventas de Technology con línea de tendencia")
plt.xlabel("Año")
plt.ylabel("Ventas ($)")
plt.legend()
plt.grid(True)
plt.show()

print(f"Crecimiento promedio por año: ${coeficientes[0]:,.2f}")
```

```python
media_movil = ventas_anuales.rolling(window=2).mean()

plt.figure(figsize=(10,5))
plt.plot(ventas_anuales.index, y, marker="o", label="Ventas reales", color="darkorange", alpha=0.6)
plt.plot(ventas_anuales.index, media_movil.values, marker="s", label="Media móvil (2 años)", color="blue", linewidth=2)
plt.plot(ventas_anuales.index, tendencia(x), linestyle="--", label="Línea de tendencia", color="red")
plt.title("Ventas de Technology: reales, media móvil y tendencia")
plt.xlabel("Año")
plt.ylabel("Ventas ($)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from prophet import Prophet

datos_prophet = ventas_anuales.reset_index()
datos_prophet.columns = ["ds", "y"]
datos_prophet["ds"] = pd.to_datetime(datos_prophet["ds"], format="%Y")

modelo = Prophet(yearly_seasonality=False)
modelo.fit(datos_prophet)

futuro = modelo.make_future_dataframe(periods=3, freq="YE")
prediccion = modelo.predict(futuro)

modelo.plot(prediccion)
plt.title("Proyección de ventas — Technology (3 años adelante)")
plt.xlabel("Año")
plt.ylabel("Ventas ($)")
plt.show()
```
